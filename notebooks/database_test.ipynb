```python
import pandas as pd
from sqlalchemy import create_engine
from dotenv import load_dotenv
import os

# Load env
load_dotenv()
engine = create_engine(f"postgresql://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}")

# Check row count
query = "SELECT COUNT(*) FROM pitch_data;"
count = pd.read_sql(query, engine)
print(count)
```

```python
df = pd.read_sql("SELECT * FROM pitch_data LIMIT 10;", engine)
df.head()
```

```python
query = """
SELECT pitcher, stand, balls, strikes, inning, outs_when_up, pitch_type
FROM pitch_data
WHERE pitch_type IS NOT NULL
"""
test_df = pd.read_sql(query, engine)

pitch_map = {
    'FF': 'Four-Seam Fastball',
    'FT': 'Two-Seam Fastball',
    'SI': 'Sinker',
    'FC': 'Cutter',
    'SL': 'Slider',
    'CH': 'Changeup',
    'CU': 'Curveball',
    'KC': 'Knuckle Curve',
    'KN': 'Knuckleball',
    'FS': 'Split-Finger Fastball',
    'ST': 'Sweeper',
    'EP': 'Eephus',
    'SC': 'Screwball',
    'FO': 'Forkball'
}

test_df['pitch_type_full'] = test_df['pitch_type'].map(pitch_map)

test_df.head(20)
```
